Práctica 2 - Limpieza de datos

```python
import pandas as pd

#creamos el dataframe
df = pd.read_csv('China_Tech_HK.csv')
df
```

```python
#cambiamos el formato de nuestros datos de fecha "trade_date"
df['fecha'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')
df
```

```python

#encontrar todos los codigos unicos de cada empresa
codigos = df['ts_code'].unique()
print(codigos)
```

```python
#cambiamos los codigos por su empresa correspondiente
df.replace("00020.HK","SenseTime",inplace=True)
df.replace("00700.HK","Tencent",inplace=True)
df.replace("00992.HK","Lenovo",inplace=True)
df.replace("01024.HK","Kuaishou",inplace=True)
df.replace("01810.HK","Xiaomi",inplace=True)
df.replace("03690.HK","Meituan",inplace=True)
df.replace("09618.HK","JD.com",inplace=True)
df.replace("09626.HK","Bilibili",inplace=True)
df.replace("09888.HK","Baidu",inplace=True)
df.replace("09961.HK","Trip.com",inplace=True)
df.replace("09988.HK","Alibaba",inplace=True)
df['amount'] = df['amount'] /1000000
```

```python
#guardamos los cambios y mostramos head
df.to_csv('China_Tech_HK.csv', index=False)
print(df.head())
```

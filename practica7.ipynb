```python
import pandas as pd

# Creamos el dataframe
df = pd.read_csv('China_Tech_HK.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Scatter plot con diferentes colores para cada stock
plt.figure(figsize=(10, 6))

for stock_code, data in df.groupby('ts_code'):
    plt.scatter(data['vol'], data['close'], label=stock_code, alpha=0.7)

plt.title('Dispersion de Precio de Stock y Volumen de Transacciones por Empresa')
plt.xlabel('Volumen de Transacciones')
plt.ylabel('Precio de Stock')
plt.legend(title='Empresa', bbox_to_anchor=(1.05, 1), loc='upper left')


plt.style.use('miestilo.mplstyle')
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300


plt.show()
```

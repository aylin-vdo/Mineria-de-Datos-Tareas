```python
import pandas as pd

# Creamos el dataframe
df = pd.read_csv('China_Tech_HK.csv')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

columnas_clustering = ['open', 'high', 'low', 'close', 'change', 'pct_chg', 'vol', 'amount']

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[columnas_clustering])

num_clusters = 3

kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)

plt.figure(figsize=(12, 8))

for cluster_num in range(num_clusters):
    cluster_data = df[df['cluster'] == cluster_num]
    plt.scatter(cluster_data['pct_chg'], cluster_data['vol'], label=f'Cluster {cluster_num}', alpha=0.7)

plt.title('KMeans Clustering de los Stocks')
plt.xlabel('Porcentaje de Cambio')
plt.ylabel('Volumen')
plt.legend()

plt.style.use('miestilo.mplstyle')
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

plt.show()
```

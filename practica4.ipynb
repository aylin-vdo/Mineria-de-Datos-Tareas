```python
import pandas as pd

# Creamos el dataframe
df = pd.read_csv('China_Tech_HK.csv')
```

```python
import matplotlib.pyplot as plt

# Calculamos la cantidad promedio por empresa
cant_prom_por_empresa = df.groupby('ts_code')['amount'].mean().reset_index()

# Describimos el tama√±o de nuestra grafica
plt.figure(figsize=(12, 6))

# Agrego mi estilo y dpi preferidos
plt.style.use('miestilo.mplstyle')
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

# Bar plot
plt.bar(cant_prom_por_empresa['ts_code'], cant_prom_por_empresa['amount'], color='#616e88') 
plt.title('Cantidad de Transacci√≥n Promedio Para Cada Empresa')
plt.xlabel('Empresa')
plt.ylabel('Cantidad de Transacci√≥n Promedio')
plt.xticks(rotation=45, ha='right') 

plt.savefig('cant_promedio_por_empresa.png', transparent=False, facecolor='#2e3440')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

company_data = df[df['ts_code'] == 'Lenovo']

company_data = company_data[::-1]

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(company_data['fecha'], company_data['close'], label='Closing Price', color='#b48ead')
plt.title('Precio de Stock de Lenovo vs Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Closing Price (HKD)')

month_locator = mdates.MonthLocator()
date_formatter = mdates.DateFormatter('%b')

plt.gca().xaxis.set_major_locator(month_locator)
plt.gca().xaxis.set_major_formatter(date_formatter)

plt.legend()
plt.grid(True)

plt.savefig('precio_stocks_lenovo.png', transparent=False, facecolor='#2e3440')
plt.show()
```

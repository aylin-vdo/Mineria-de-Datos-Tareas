Práctica 6 - Regresion Lineal y Correlación

```python
import pandas as pd

# Creamos el dataframe
df = pd.read_csv('China_Tech_HK.csv')
```

```python
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt

stock_seleccionado = 'Lenovo'
datos_seleccionados = df[df['ts_code'] == stock_seleccionado]

# Linear Regression
slope, intercept, r_value, p_value, std_err = linregress(datos_seleccionados['vol'], datos_seleccionados['close'])

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
plt.scatter(datos_seleccionados['vol'], datos_seleccionados['close'], label='Data points', color='#616e88')
plt.plot(datos_seleccionados['vol'], intercept + slope * datos_seleccionados['vol'], color='#87bfcf', label='Regression line')
plt.title(f'Regresion linear para {stock_seleccionado}')
plt.xlabel('Volumen de Transacciones')
plt.ylabel('Precio de Stock')
plt.legend()

plt.style.use('miestilo.mplstyle')
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300


plt.show()
```
